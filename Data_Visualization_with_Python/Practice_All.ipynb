```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_csv(r'DataSets/CanadianImmigration.csv')
df_can.head()
```

```python
df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV', 'DevName'], axis=1, inplace=True)
df_can.head()
```

```python
df_can.rename({'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, axis=1, inplace=True)
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
df_can.index.name = None
df_can.head()
```

```python
plt.style.use(['ggplot'])
```

```python
years = list(map(str, range(1980, 2014, 1)))
df_can['Total'] = df_can[years].sum(axis=1)
df_can.head()
```

```python
df_5top = df_can.sort_values(by='Total', ascending=False, axis=0)
df_line = df_5top[years].head().T
df_line.index.map(int)
```

```python
df_line.plot(kind='line', figsize=(14, 8))
plt.ylabel('Numbers of immigrants')
plt.xlabel('Years')
plt.title('Numbers of immigrants top 5 countries from 1980 to 2013')
# plt.savefig(fname=r'Save_Image\top_5_countries.png')
plt.show()
```

```python
df_line.plot(kind='area', alpha=0.6, figsize=(14, 8))
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.title('Numbers of immigrants top 5 countries from 1980 to 2013')
```
